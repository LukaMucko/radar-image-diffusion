```python
def plot_images(images):
    fig, axes = plt.subplots(4, 4, figsize=(8, 8))

    # Plot each image in the grid
    for i, ax in enumerate(axes.flatten()):
        ax.imshow(images[i], cmap='gray')
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```
